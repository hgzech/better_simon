```python
# Extracting necessary columns and converting to list of dicts
trials_list = df.drop(columns=['S2', 'R2','S1_R1_corresp','S2_R2_corresp']).to_dict(orient='records')

# Formatting each dictionary in the list
formatted_trials = []
for trial in trials_list:
    formatted_trial = {
        'S1': trial['S1'],
        'R1': trial['R1'],
        'info': ', '.join(f'{k}: {v}' for k, v in trial.items() if k not in ['S1', 'R1'])
    }
    formatted_trials.append(formatted_trial)

formatted_trials
```
